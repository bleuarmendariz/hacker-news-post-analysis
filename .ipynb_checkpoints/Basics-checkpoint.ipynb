# Hacker News Post Engagement - Comparing 'Ask HN' to 'Show HN'

In this project we are going to determine what types of posts recieve the most engagement on hackernews.

We will guage 'engagement' as comments. Therefore we have removed any posts with no comments.

There are two main questions:

> Which type post shows more engagement?

> Does time of posting effect engagement?

[The dataset can be found here](https://www.kaggle.com/hacker-news/hacker-news-posts)

### Read in data

```python
from csv import reader
open_file = open('hacker_news.csv')
read_file = reader(open_file)
hn = list(read_file)
```

```python
print(hn[:6])
```

```python
headers = hn[1]
hn = hn[1:]
hn[:5]
```

### Comparing Ask Posts to Show posts in Hacker news - Which posts show more engagement?

Seperating our lists between AskHN and ShowHN

```python
asks_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    
    if title.lower().startswith('ask hn'):
        asks_posts.append(row)
    elif title.lower().startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)

print('There are {} that are Ask HN'.format(len(asks_posts)))
print('There are {} that are Show HN'.format(len(show_posts)))
print('There are {} other posts'.format(len(other_posts)))
```

```python
asks_posts[:5]
```

```python
show_posts[:5]
```

```python
total_ask_comments = 0 #counter

for row in asks_posts:
    num_comments = int(row[4])
    total_ask_comments += num_comments

avg_ask_comments = total_ask_comments / len(asks_posts)

print('Our total Ask HN Comments: {}'.format(total_ask_comments))
print('Our Average comments per post: {}'.format(avg_ask_comments))
```

```python
total_show_comments = 0 #counter

for row in show_posts:
    num_comments = int(row[4])
    total_show_comments += num_comments

avg_show_comments = total_show_comments / len(show_posts)

print('Our total Show HN Comments: {}'.format(total_show_comments))
print('Our Average comments per post: {}'.format(avg_show_comments))
```

### Ask vs Show Engagement Results

We can see that Ask HN has double the comments, and .5x the amount of average comments per post. The simple explanation is that asking a question is a call to action (kind of). Show HN does not give a call to action, thus less popularity and engagement. Show HN is less interactive, and doesnt inspire a discussion or comment.

### Focusing on Ask Posts - Does time of posting effect engagement?

Does time of day have an impact on post engagement?

To find this out we will need to:

1. Calculate the number of ask posts created in each hour of the day, along with the number of comments recieved

2. Calculate the average number of comments ask posts recieve by hour created.

Getting posts per hour, and comments per hour into dictionaries

```python
import datetime as dt
result_list = []

for row in asks_posts:
    created_at = row[6]
    num_comments = row[4]
    
    result_list.append([created_at,num_comments])

counts_by_hour = {}
comments_by_hour = {}
date_format = "%m/%d/%Y %H:%M"
for row in result_list:
    hour = row[0]
    comment = int(row[1])
    hour_dt = dt.datetime.strptime(hour, date_format).strftime('%H')
    
    if hour_dt not in counts_by_hour:
        counts_by_hour[hour_dt] = 1
        comments_by_hour[hour_dt] = comment
    elif hour_dt in counts_by_hour:
        counts_by_hour[hour_dt] += 1
        comments_by_hour[hour_dt] += comment
```

Finding the average per hour

```python
avg_by_hour = []

# hour, count, total
for hour in counts_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour] / counts_by_hour[hour]])
avg_by_hour
```

Swapping the indexes so we can sort from highest to lowest

```python
swap_avg_by_hour = []

for i in avg_by_hour:
    swap_avg_by_hour.append([i[1], i[0]])
swap_avg_by_hour    
```

Getting the top 5 values

```python
sort_avg_list = sorted(swap_avg_by_hour, reverse = True)
sort_avg_list[:5]
```

The top 5 times with highest average comments are:
> 3pm - With 38 average comments per post

> 2am - With 23 average comments per post

> 8pm - With 21 average comments per post

> 4th and 5th tied at 4pm - With 16 comments per post

## Conclusion

Ask HN posts get more attention, and have more frequency on the site. Time of day **does** have a significant impact on post engagement, with the top 5 times being later in the day, from 3pm(the highest average engagement) to 8pm.
